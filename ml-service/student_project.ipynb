```python
from sklearn.model_selection import train_test_split , GridSearchCV #train_test_split for splitting our data in training and testing sets , Grid search cross validation on parameter tuning  
from sklearn.preprocessing import LabelEncoder #for transforming the categorial features into numeric ones  
from sklearn.metrics import mean_squared_error , r2_score #mean square and R square score on model evolation 
from sklearn.linear_model import LinearRegression  #modelling Linear Regression + regularization.Prevents overfitting by penalizing large coefficients.Better when features are correlated.
from sklearn.ensemble import RandomForestRegressor #modelling  It:Builds many decision treesAverages results,Reduces overfitting,Usually performs very well.
from sklearn.tree import DecisionTreeRegressor #modelling Simple tree-based model.Splits data into branches,Easy to visualize,Can overfit easily
```

```python
df.columns
```

```python
df.head(2)
```

```python
features = ["study_hours_per_day" , "attendance_percentage" ,"mental_health_rating" , "sleep_hours" , "part_time_job" ] #features = on which basis you want to predict the target
```

```python
target = "exam_score"   #target = what you want to predict
```

```python
df_model = df[features + [target]].copy()  #features is already a list , [target] converts the string "exam_score" into a list , what it means is ki tumne features vali list me bs exam_score add kiya

# Without .copy():
# df_model may still be linked to original df.
# If you modify df_model, it could modify original data.
# .copy() creates a separate independent dataframe.
# Safe practice.

```

```python
df_model
```

```python
le = LabelEncoder()
```

```python
df_model["part_time_job"] =le.fit_transform(df_model["part_time_job"]) # 0 for no and 1 for yes , now if you call df_model in part time job this will take place
```

```python
X = df_model[features]
```

```python
Y = df_model[target]
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y, test_size=0.2) #0.2 means we are going to keep 20% in the testing set
```

```python
len(Y_test)
```

```python
len(Y_train)
```

```python
models = {
    "LinearRegression": {
        "model" : LinearRegression(),
        "params": {} 
    },
    "DecisionTree" : {
        "model" : DecisionTreeRegressor() , 
        "params" : {"max_depth" : [3,5,10] , "min_samples_split" : [2,5]}
    }, 
    "RandomForest" : {
        "model" : RandomForestRegressor(),
        "params" : {'n_estimators' : [50,100] , "max_depth":[5,10]}
    }    
}
```

```python
best_models =[]
```

```python
for name , config in models.items():
    print(f"Training {name}")

    grid = GridSearchCV(config["model"] , config["params"] , cv=5 , scoring= "neg_mean_squared_error") #config = settings 
    #You are saying:
    # Take this model
    # Try all parameter combinations
    # Use cross-validation
    # Choose best one using error metric
    grid.fit(X_train , Y_train) 

    y_pred = grid.predict(X_test)
    rmse = np.sqrt(mean_squared_error(Y_test , y_pred))
    r2 = r2_score(Y_test , y_pred)
    best_models.append({
        "model": name,
        "best_params" : grid.best_params_, 
        "rmse" : rmse,
        "r2" : r2 
    })
```

```python
best_models
```

```python
result_df = pd.DataFrame(best_models)
```

```python
result_df.sort_values(by= "rmse")
```

```python
import joblib #for model exporting

best_row = result_df.sort_values(by="rmse").iloc[0]
```

```python
best_row
```

```python
best_model_name = best_row["model"]
```

```python
best_model_name
```

```python
best_model_config = models[best_model_name]
```

```python
best_model_config
```

```python
final_model = best_model_config["model"].set_params(**best_row["best_params"])
```

```python
final_model.fit(X,Y)
```

```python
final_model.predict(X_test)
```

```python
joblib.dump(final_model , "best_model.pkl")
```
